```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load grayscale image (use your low-contrast image here)
img_path = r'Ass5_.jpg'   # replace with low contrast image
img_gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
if img_gray is None:
    raise ValueError("Image not found!")

# --- 1. Contrast Stretching ---
# find min and max intensity values
min_val, max_val = np.min(img_gray), np.max(img_gray)

# apply contrast stretching formula
stretched_img = ((img_gray - min_val) / (max_val - min_val)) * 255
stretched_img = stretched_img.astype(np.uint8)

# --- 2. Histogram Equalization ---
equalized_img = cv2.equalizeHist(img_gray)

# --- 3. Plot images ---
plt.figure(figsize=(15,8))

# Original
plt.subplot(3,3,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(3,3,2)
plt.imshow(stretched_img, cmap='gray')
plt.title("Contrast Stretched")
plt.axis('off')

plt.subplot(3,3,3)
plt.imshow(equalized_img, cmap='gray')
plt.title("Histogram Equalized")
plt.axis('off')

# --- 4. Histograms ---
plt.subplot(3,3,4)
plt.hist(img_gray.ravel(), bins=256, range=[0,256], color='black')
plt.title("Original Histogram")

plt.subplot(3,3,5)
plt.hist(stretched_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Contrast Stretched Histogram")

plt.subplot(3,3,6)
plt.hist(equalized_img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Equalized Histogram")

plt.tight_layout()
plt.show()
```
